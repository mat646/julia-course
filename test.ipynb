```julia
using Plots
using DataFrames
using DifferentialEquations
```

```julia
function lorenz(du,u,p,t)
 du[1] = 10.0*(u[2]-u[1])
 du[2] = u[1]*(28.0-u[3]) - u[2]
 du[3] = u[1]*u[2] - (8/3)*u[3]
end
```

```julia
u0 = [1.0;0.0;0.0]
tspan = (0.0,100.0)
prob = ODEProblem(lorenz,u0,tspan)
sol = solve(prob)
```

```julia
plot(sol,vars=(1,2,3))
```

```julia
function LVE(du,u,p,t)
 du[1] = u[1] - u[2]*u[1]
 du[2] = -u[2] + u[1]*u[2]
end
```

```julia
u0 = [5.0;3.0]
tspan = (0.0,40.0)
problem = ODEProblem(LVE, u0, tspan)
sol = solve(problem, RK4())
```

```julia

plot(sol,vars=(1))
plot!(sol,vars=(2))
```
